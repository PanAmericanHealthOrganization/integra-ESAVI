```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
## datos pora una conexión local
db_user = "dhis"
db_password = "dhis"
db_host = "localhost"  # Or your PostgreSQL server's IP address/hostname
db_port = "5432"       # Default PostgreSQL port
db_name = "dhi_esavi"
```

```python
connection_string = f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}"

# Create the SQLAlchemy engine
engine = create_engine(connection_string)
# Test the connection
try:
    connection = engine.connect()
    print("Conexión exitosa a la base de datos")
    connection.close()
except Exception as e:
    print(f"Error al conectar a la base de datos: {e}")
    raise e
```

```python
df_notificacion = pd.read_sql('SELECT * FROM dhi_esavi."TR_NOTIFICACION"', engine).astype(str)
df_notificacion = df_notificacion.loc[:, ~df_notificacion.columns.str.startswith('AUD_')]
#
df_catalogo = pd.read_sql('SELECT * FROM dhi_esavi."TC_CATALOGO"', engine).astype(str)
df_catalogo = df_catalogo.loc[:, ~df_catalogo.columns.str.startswith('AUD_')]
#
df_grupo_etario = pd.read_sql('SELECT * FROM dhi_esavi."TC_GRUPO_ETARIO"', engine).astype(str)
df_grupo_etario = df_grupo_etario.loc[:, ~df_grupo_etario.columns.str.startswith('AUD_')]
#
df_parametro = pd.read_sql('SELECT * FROM dhi_esavi."TC_PARAMETRO"', engine).astype(str)
df_parametro = df_parametro.loc[:, ~df_parametro.columns.str.startswith('AUD_')]
#
df_tipo_catalogo = pd.read_sql('SELECT * FROM dhi_esavi."TC_TIPO_CATALOGO"', engine).astype(str)
df_tipo_catalogo = df_tipo_catalogo.loc[:, ~df_tipo_catalogo.columns.str.startswith('AUD_')]
#
df_gaceta = pd.read_sql('SELECT * FROM dhi_esavi."TG_GACETA"', engine).astype(str)
df_gaceta = df_gaceta.loc[:, ~df_gaceta.columns.str.startswith('AUD_')]
#
df_antecedentes_embarazo = pd.read_sql('SELECT * FROM dhi_esavi."TR_ANTECEDENTES_EMBARAZO"', engine).astype(str)
df_antecedentes_embarazo = df_antecedentes_embarazo.loc[:, ~df_antecedentes_embarazo.columns.str.startswith('AUD_')]    
#
df_antecedentes_enfermedades_previas = pd.read_sql('SELECT * FROM dhi_esavi."TR_ANTECEDENTES_ENFERMEDADES_PREVIAS"', engine).astype(str)    
df_antecedentes_enfermedades_previas = df_antecedentes_enfermedades_previas.loc[:, ~df_antecedentes_enfermedades_previas.columns.str.startswith('AUD_')]
#
df_antecedentes_evento_adverso = pd.read_sql('SELECT * FROM dhi_esavi."TR_ANTECEDENTES_EVENTO_ADVERSO"', engine).astype(str)    
df_antecedentes_evento_adverso = df_antecedentes_evento_adverso.loc[:, ~df_antecedentes_evento_adverso.columns.str.startswith('AUD_')]
#
df_antecedentes_medico = pd.read_sql('SELECT * FROM dhi_esavi."TR_ANTECEDENTES_MEDICO"', engine).astype(str)
df_antecedentes_medico = df_antecedentes_medico.loc[:, ~df_antecedentes_medico.columns.str.startswith('AUD_')]
#
df_causalidad_esavi = pd.read_sql('SELECT * FROM dhi_esavi."TR_CAUSALIDAD_ESAVI"', engine).astype(str)
df_causalidad_esavi = df_causalidad_esavi.loc[:, ~df_causalidad_esavi.columns.str.startswith('AUD_')]
#
df_datos_esavi = pd.read_sql('SELECT * FROM dhi_esavi."TR_DATOS_ESAVI"', engine).astype(str)
df_datos_esavi = df_datos_esavi.loc[:, ~df_datos_esavi.columns.str.startswith('AUD_')]
#
df_dato_vacuna = pd.read_sql('SELECT * FROM dhi_esavi."TR_DATO_VACUNA"', engine).astype(str)
df_dato_vacuna = df_dato_vacuna.loc[:, ~df_dato_vacuna.columns.str.startswith('AUD_')]
#
df_dato_vacunacion = pd.read_sql('SELECT * FROM dhi_esavi."TR_DATO_VACUNACION"', engine).astype(str)
df_dato_vacunacion = df_dato_vacunacion.loc[:, ~df_dato_vacunacion.columns.str.startswith('AUD_')]
#
df_desenlace_esavi = pd.read_sql('SELECT * FROM dhi_esavi."TR_DESENLACE_ESAVI"', engine).astype(str)
df_desenlace_esavi = df_desenlace_esavi.loc[:, ~df_desenlace_esavi.columns.str.startswith('AUD_')]
#
df_esavi_durante_embarazo = pd.read_sql('SELECT * FROM dhi_esavi."TR_ESAVI_DURANTE_EMBARAZO"', engine).astype(str)
df_esavi_durante_embarazo = df_esavi_durante_embarazo.loc[:, ~df_esavi_durante_embarazo.columns.str.startswith('AUD_')]
#
df_establecimiento = pd.read_sql('SELECT * FROM dhi_esavi."TR_ESTABLECIMIENTO"', engine).astype(str)
df_establecimiento = df_establecimiento.loc[:, ~df_establecimiento.columns.str.startswith('AUD_')]
#
df_gravedad_esavi = pd.read_sql('SELECT * FROM dhi_esavi."TR_GRAVEDAD_ESAVI"', engine).astype(str)
df_gravedad_esavi = df_gravedad_esavi.loc[:, ~df_gravedad_esavi.columns.str.startswith('AUD_')]
#
df_investigacion = pd.read_sql('SELECT * FROM dhi_esavi."TR_INVESTIGACION"', engine).astype(str)    
df_investigacion = df_investigacion.loc[:, ~df_investigacion.columns.str.startswith('AUD_')]
#
df_medicamento = pd.read_sql('SELECT * FROM dhi_esavi."TR_MEDICAMENTO"', engine).astype(str)
df_medicamento = df_medicamento.loc[:, ~df_medicamento.columns.str.startswith('AUD_')]
#
df_paciente = pd.read_sql('SELECT * FROM dhi_esavi."TR_PACIENTE"', engine).astype(str)
df_paciente = df_paciente.loc[:, ~df_paciente.columns.str.startswith('AUD_')]
#
df_paciente_embarazada = pd.read_sql('SELECT * FROM dhi_esavi."TR_PACIENTE_EMBARAZADA"', engine).astype(str)
df_paciente_embarazada = df_paciente_embarazada.loc[:, ~df_paciente_embarazada.columns.str.startswith('AUD_')]
#
df_sync_process = pd.read_sql('SELECT * FROM dhi_esavi."TR_SYNC_PROCESS"', engine).astype(str)
df_sync_process = df_sync_process.loc[:, ~df_sync_process.columns.str.startswith('AUD_')]
#
df_vacunometro = pd.read_sql('SELECT * FROM dhi_esavi."TR_VACUNOMETRO"', engine).astype(str)
df_vacunometro = df_vacunometro.loc[:, ~df_vacunometro.columns.str.startswith('AUD_')]
#
```
